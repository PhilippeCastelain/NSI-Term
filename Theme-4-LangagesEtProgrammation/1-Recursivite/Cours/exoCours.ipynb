```python
from timeit import default_timer as timer
from random import randint

def somme(liste):
    S = 0
    for e in liste:
        S = S +e
    return S

def sommeliste(liste):
    if len(liste) == 1:
        return liste[0]
    else :
        return liste[0] + sommeliste(liste[1:])

L = [randint(0,100) for i in range(300)]
    
debut = timer()
print(sommeliste(L))
fin = timer()
print(fin - debut)

debut = timer()
print(somme(L))
fin = timer()
print(fin - debut)
```
